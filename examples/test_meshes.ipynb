```julia

function dipoles2(g::CartesianGrid, s::AbstractTarget)
    coords = SVector{3,Float64}[]
    inds = CartesianIndices[]
    for ind in CartesianIndices(g)
        coord = g[ind]
        if isinside(coord..., s)
            push!(coords, coord)
            push!(inds, ind)
        end
    end
    return coords,inds
end


function discretize(g::CartesianGrid, s::AbstractTarget)

    eps = 1
    out = zeros(typeof(eps), size(g))

    inds = CartesianIndices(g)
    for ind in inds
        coord = g[ind]
        if isinside(coord..., s)
            out[ind] = eps
        end
    end
    return out
end
```

```julia
# PROBLEM: eltype(typeof(grid))
# PROBLEM: ndims(typeof(grid))
# PROBLEM: eltype(vertices(grid))
# PROBLEM: CartesianIndices(grid) 


coordinate(grid::CartesianGrid, ind::CartesianIndex) = grid.origin + (ind.I .- 1) .* grid.spacing 

function dipoles(grid::CartesianGrid, target::Primitive)
    N, T = typeof(grid).parameters

    coords = Point{N, T}[]
    inds = CartesianIndex{N}[]
    for ind in CartesianIndices(grid.dims)
        coord = coordinate(grid, ind)
        if coord in target
            push!(coords, coord)
            push!(inds, ind)
        end
    end
    return coords,inds
end
```

```julia
coords, inds = dipoles(grid, target);
```

```julia
plot(coords; color=1:length(coords))
```
